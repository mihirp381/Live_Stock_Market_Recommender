```python
##I need the pre-open prices stored in one and I need the current price at 9:17 and then gimme the most occuring change
pr_comp= ['Tata Steel', 'Cummins', 'Bata India', 'Cadila Health', 'Eicher Motors', 'Power Grid Corp']
pr_op= pd.DataFrame({})
pr_op['Name']= pr_comp
pr_op['Pre-open']= np.zeros(len(pr_comp))
pr_op['Current-Price']= np.zeros(len(pr_comp))
pr_op['% Change']= np.zeros(len(pr_comp))

def pre_open(company_list: list):
   
    while True:
        
        for i in company_list:
            try:
                link = companies[i]
                page_= requests.get(link)
                bsobj_= bs4.BeautifulSoup(page_.text, 'html.parser')
                price= bsobj_.find('body', {'class' : 'container1280'}).find('div', {'class': 'pnc_wrapper'}).find('div', {'id': 'mainprice'}).find('div', {'class': 'tab-content'}).find('div', {'class': 'inprice1 nsecp'}).text
                pre_open = float(price.replace(",", ""))

                pr_op.loc[ pr_op['Name'] == i, ['Pre-open'] ]= pre_open
            except:
                pass
        
        time.sleep(300)
        
        for i in company_list:
            try:
                link = companies[i]
                page_= requests.get(link)
                bsobj_= bs4.BeautifulSoup(page_.text, 'html.parser')
                price = bsobj.find('body', {'class' : 'container1280'}).find('div', {'class': 'pnc_wrapper'}).find('div', {'id': 'mainprice'}).find('div', {'class': 'tab-content'}).find('div', {'class': 'inprice1 nsecp'}).text #.find('div', {'class': 'tab-pane fade active'}).find('div', {'class': 'clearfix'}).find('div', {'class': 'inindi_price'}).find('div', {'class': 'clearfix mb10'}).find('div', {'class': 'indimprice'}).find('div', {'class': 'inprice1 nsecp'}).text
                curr_price = float(price.replace(",", ""))

                pr_op.loc[ pr_op['Name'] == i, ['Current-Price'] ]= curr_price
            except:
                pass
        
        for c in company_list:
            last, current= pr_op.loc[ pr_op['Name'] == c]['Pre-open'] , pr_op.loc[ pr_op['Name'] == c]['Current-Price'] 
            pr_op.loc[ pr_op['Name'] == c, ['% Change']]= ((current-last)/last)*100
        
        text_file = open("pre-open.html", "w")
        text_file.write(pr_op.to_html())
        
        text_file.close()
        
        t = time.localtime()
        current_time = time.strftime("%H:%M:%S", t)
        
        print("Last updated at: "+ str(current_time) )
```
